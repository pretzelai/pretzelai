```python
%%html
<style>
    .jp-Notebook-cell { padding-left: 100px; }
    .jp-Notebook-cell::after { content: 'CSS ::after element'; }
</style>
```

```python
%%html
<script type="module">
setInterval(() => {
    const injected = document.body.querySelectorAll('[data-origin="html"]');
    for(const span of injected) {
        span.remove();
    }
    const cells = document.body.querySelectorAll('.jp-Notebook-cell');
    for(const cell of cells) {
        cell.insertAdjacentHTML('beforeend', '<span data-origin="html">JavaScript injected from HTML</span>')
    }
}, 300)
</script>
```

```python
%%javascript
setInterval(() => {
    const injected = document.body.querySelectorAll('[data-origin="javascript"]');
    for(const span of injected) {
        span.remove();
    }
    const cells = document.body.querySelectorAll('.jp-Notebook-cell');
    for(const cell of cells) {
        cell.insertAdjacentHTML('afterbegin', '<span data-origin="javascript">JavaScript injected header</span>')
    }
}, 300)
```
